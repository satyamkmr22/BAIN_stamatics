 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python
def P(event_condition, outcomes, probabilities):
    event_prob = 0
    for outcome, probability in zip(outcomes, probabilities):
        if event_condition(outcome):
            event_prob += probability
    return event_prob
outcomes = ["Strawberry", "Cinnamon", "BlueBerry"]
probabilities = [0.5, 0.25, 0.25]
event_A = lambda x: x == "Strawberry"
event_B = lambda x: x != "Cinnamon"
prob_A = P(event_A, outcomes, probabilities)
prob_B = P(event_B, outcomes, probabilities)

print(f"Probability of event A (Strawberry): {prob_A}")
print(f"Probability of event B (Not Cinnamon): {prob_B}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, betabinom

gaussian = norm(loc=0, scale=1)
betabinomial = betabinom(n=10, a=2, b=2)

probabilities = np.linspace(0, 1, 100)
gaussian_ppf = gaussian.ppf(probabilities)

betabinomial_ppf = betabinomial.ppf(probabilities)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(probabilities, gaussian_ppf, label='Gaussian PPF')
plt.xlabel('Probability')
plt.ylabel('Value')
plt.title('PPF of Gaussian Distribution')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(probabilities, betabinomial_ppf, label='BetaBinomial PPF', color='orange')
plt.xlabel('Probability')
plt.ylabel('Value')
plt.title('PPF of BetaBinomial Distribution')
plt.legend()

plt.tight_layout()
plt.show()

```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

a = 5
skewnormal = skewnorm(a)

theoretical_mean = skewnormal.mean()
theoretical_variance = skewnormal.var()

print(f"Theoretical Mean: {theoretical_mean}")
print(f"Theoretical Variance: {theoretical_variance}")

def compute_sample_moments(distribution, sample_size):
    samples = distribution.rvs(size=sample_size)
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples, ddof=1)
    return sample_mean, sample_variance

sample_sizes = [10, 100, 1000]
for size in sample_sizes:
    sample_mean, sample_variance = compute_sample_moments(skewnormal, size)
    print(f"\nSample Size: {size}")
    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Variance: {sample_variance}")
```
